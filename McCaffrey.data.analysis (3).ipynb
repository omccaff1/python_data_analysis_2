```python
#Let's give a more descriptive name to the dataframe.
mobile_homes = pd.read_csv('../Data/ROC_data.csv')
```

```python
#Let's get a quick summary of the data.
mobile_homes.describe()
```

```python
#Let's sort the data by 'year_joined' to see which communities have been resident-owned for the longest.
mobile_homes.sort_values(by='year_joined').head(10)
```

```python
#Let's grab a single column and print the first 10 values.
mobile_homes.state.head(10)
```

```python
#Let's take the "number_homes" column. What are the smallest and largest communities on our list?
mobile_homes.number_homes.min(),mobile_homes.number_homes.max()
```

```python
#Which states have the most resident-owned mobile home communities?
#Let's count the number per state by using 'value_counts'
mobile_homes['state'].value_counts()
```

```python
#How many communities were added each year until 2018? Is the number growing?
mobile_homes.groupby('year_joined').count()
```
